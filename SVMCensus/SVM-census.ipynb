```python
import pandas as pd
import numpy as np
```

```python
original_data =pd.read_csv("data/adult.csv", names=["Age", "Workclass","fnlwgt", "Education","Education-num", "martial Sum","Occupation", "Relationship"],
sep=r'\s*,\s*',
engine ='python',
na_values="?")

original_data.head()
```

```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
fig =plt.figure(figsize=(20,20))
cols =3
rows = math.ceil(float(original_data.shape[1] /cols))

for i , column in enumerate(['Age',"Workclass","fnlwgt", "Education","Occupation"]):
    ax= fig.add_subplot(rows,cols,i+1)
    ax.set_title(column)
    if original_data.dtypes[column] ==np.object:
        original_data[column].value_counts().plot(kind="bar", axes=ax)
    else:
        original_data[column].hist(axes=ax)
        plt.xticks(rotation="vertical")
plt.subplots_adjust(hspace=0.7 ,wspace=0.2)
plt.show()
        
```

```python
import sklearn.preprocessing as preprocessing
le = preprocessing.LabelEncoder()
original_data['Occupation'] =le.fit_transform(original_data['Occupation'].astype(str))
original_data.head()
```

```python
original_data['Education'] =le.fit_transform(original_data['Education'].astype(str))
original_data.head()
```

```python
original_data.Education.unique()
```

```python
original_data.groupby('Education').Education.mean().plot(kind='bar')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X= original_data[['Education', 'Occupation']]
y= original_data[['Education']]
```

```python
X_train,x_test, y_train, y_test =train_test_split(X ,y,test_size=0.2, random_state=0)
```

```python
from sklearn.svm import SVC
#Declaring the SVC with no tuning 
classifier =SVC()
# Fiting the data .This is where the SVM will learn
classifier.fit(X_train,y_train)
#Predicting the result and giving the accuracy
score =classifier.score(x_test, y_test)
```

```python
print(score)
```
